```python
pp=lambda a:np.array(a); r=range; import os; import seaborn as sns
import numpy as np; import pandas as pd; from collections import Counter; import matplotlib.pyplot as plt; 转整=lambda a:pp(a).astype(int); 变幅=lambda 次,首:(次-首)/首*100
import matplotlib.font_manager as f; f.fontManager.addfont("s.ttf"); plt.rcParams['font.sans-serif']=[f.FontProperties(fname="s.ttf").get_name()]; plt.rcParams['axes.unicode_minus']=0
def 年周分析(价矢,参年,期数,步距): # 幂是步距较单月较步矩的倍数,价矢索引最值'期数*参年'最多取到年末素如10*12,因py首数为0即末比是去12月和今1间的,再后则作测试,点组中期为列矢而年为行矢
    点组=np.arange(期数).reshape(-1,1)+期数*np.arange(参年).reshape(1,-1); 点年矩阵=((价矢[点组+步距]/价矢[点组])**(期数/12/步距)-1)*100; w=pp([1**i for i in r(参年)])
    return 转整(np.average(点年矩阵,axis=1,weights=w/np.sum(w))), 点年矩阵.astype(int) # 各月收益和盈率,1表跨列即横向操作,w即未归一的权序 # 建议参数设法是: 参年=5; 收益位=3; 盈率位=6; 幂指=1.3
def 矩阵对齐(矩,准行=3): 果矩=矩[矩[:,0].astype(float).argsort()[::-1]][:30]; return np.vstack([果矩,np.full((准行-len(矩),参年+2),np.nan)]) if len(矩)<准行 else 果矩
def 邻点数列生成(点数列,周期): # 主干定义过的变量可以在子函中直接用 
    for 当点 in r(1,点数列[-1],周期): # 将矩阵a中所有为负的元素所在的行全置为零得到矩阵b
        for i in r(df.shape[1]-3): # 真实首日或次日一下就可以超过首日伪数时,必大之的首点数兼作左右点,左矢若等目标点右矢亦归回
            if 点数列[i]>当点: 左点,右点=(i,i) if i==0 else (i-1,i if 点数列[i-1]!=当点 else i-1); break # 构建简写↓
        左点列.append(左点);右点列.append(右点);左=点数列[左点];右=点数列[右点];重=左点==右点;右权列.append([(当点-左)/(右-左+重),0.5][重]);左权列.append([(右-当点)/(右-左+重),0.5][重]) 
# ---------------------------------------------------------------------生成插值文件-------------------------------------------------------------- # 
周期=5; 起年=21; 入名="w216.csv" 
if not os.path.exists('五天价.csv'): 
    df = pd.read_csv(入名,encoding='utf-8'); 序矢=[(int(年)-2000-起年)*360+(int(月)-1)*30+int(日) for 素 in df.columns[3:] for 年,月,日 in [素.split("-")]] 
    行信息 = pd.DataFrame(np.array(df.iloc[:,:3]),columns=["序号","代码","名称"]); 左点列,右点列,左权列,右权列=[],[],[],[]; 邻点数列生成(序矢,周期) # 用csv首行日期轴生成必要的插值信息 
    原矩 = pp(df.iloc[:,3:df.shape[1]]); 插值矩 = 原矩[:,左点列]*左权列+原矩[:,右点列]*右权列 # 原csv标题行先从日期转为序号,根据细粒度需求设周期为5/10/15等30的因数 ↑↑
    print(df.columns[3:30],"\n序矢:",pp(序矢)[:9],"\n左点:",左点列[:9],"\n右点:",右点列[:9],"\n左权:",左权列[:9],"\n右权:",右权列[:9],"\n","原矩:",原矩.shape,"插矩:",插值矩.shape) 
    日期列名 = [f'{int((i*周期+1)/360)+起年}-{int((i*周期+1)%360/30)+1}-{(i*周期+1)%360%30}({i*周期+1})' for i in r(len(左点列))] # 价格限长并加入该列名 
    pd.DataFrame(pd.concat([行信息,pd.DataFrame(插值矩.round(2),columns=日期列名)],axis=1)).to_csv("五天价.csv",index=0) # 最后拼接两矩阵存为csv 
# ----------------------------------------------------------------------生成周期图---------------------------------------------------------------- # 
入名='五天价.csv'; 当前年=26; 前瞻年=26; 总年=5; 截止月=1; 期数=72; 准行=10
验股数=2000; 步距=6; 测否=0; 参年=5; 收益位=5; 盈率位=5; 上限=10 # 参数设置区单独抠出片空地来即可,不用放末尾套叠主函,若不显眼则用空段来指示
周期月表,测益集,周股集=[],[],[]; l=pd.read_csv(入名,encoding='utf-8'); 股号=pp(l)[:,0]; 股名=pp(l)[:,2]; 股价=pp(l)[:,3:] # 编码请试'utf-8'或'gbk',股码因重效由股名替代
期=r(期数); 截止月=截止月 if 前瞻年==当前年 else 期数; p=期数/12; 月份=[((i//p+1)+(i%p*(30/p)+1)*0.01) for i in 期]; 月周股=[[] for i in 期] # 期数非12则生成小数作为月份的日号
for i in r(实股数:=min(验股数,len(股号))): # 实验股数受总股数的限幅;月周股内部元素为列表,先将'月份'作为屋基,随后填充股名收益等字符信息,各股亦然
    当名=0; 当价=股价[i,期数*(总年-参年-(当前年-前瞻年)):]; 各月收益,各月实况=年周分析(当价,参年,期数,步距); 日号条件=1 
    for j in r(期数): # 次月收益和上月收益为与步距可能非一的当月收益及各月收益呼应,因此都叠加步距计算 
        # if 股名[i] != '海欣食品': continue #--------若想只测试一只股则取消本句注释,下面的日号有1,6,11,16,21和26 #日号条件=round((月份[j]-int(月份[j]))*100)==1 
        if np.min(各月实况[j])>=0 and 各月收益[j]>=收益位 and 日号条件: # 一年多处周期点会重复打印,故需保证与前一股码不同,先打出头年和本年数以便测试 
            周期月=月份[j]; 周股集.append(股名[i]); 周期月表.append(周期月); 月周股[j].append(np.concatenate(([各月收益[j]],各月实况[j],[股名[i]]))); 测件=测否 and 周期月<截止月 
            if 测件: 基=参年*期数+j; 上益=变幅(当价[基],当价[基-步距]); 本益=变幅(当价[基+步距],当价[基]); 测益集.append([0,round(本益,2)][上益<上限]) # print(周期月,本益,上益)
            # if 股名[i]!=当名: print(f"{i+1}.{股名[i]}"); [print(f"{前瞻年-参年+i}年收价: {[f'{当价[期数*i+j]:.1f}' for j in 期]}") for i in r(1)] # r中'参年'改1仅输头年
            # if 股名[i]!=当名: 当名=股名[i]; print(f"{前瞻年}年收价:{[f'{当价[参年*期数+j]:.2f}' for j in r(截止月*int(期数/12))]}",f"\n全年收益:{各月收益}\n")
            # if 测件 and 上益<上限: print(f"{i+1}.{股名[i]}{"="*20}{周期月}月收益{各月收益[j]}%----实收{本益:.2f}%") # 截止月需是过完整月旬后得价,否则取次末↑ 
if 测否: c=pp(测益集); c=c[c!=0]; print(f"余量测得利次{len(c)},盈占{((c>0).sum()/len(c)*100):.0f}%-{c.sum()/len(c):.2f}%") # 限制下摆自动过滤限高上益时产生的无效输出 
print(f"本轮{实股数}股中有{len(周期月表)}周期点,{len(set(周股集))}周期股"); 降频表=list(zip(*Counter(周期月表).most_common())); print(f'月份降频排:{降频表[0]}\n对应频次为:{降频表[1]}')
齐次周股 = [矩阵对齐(pp(矩),准行) if len(矩)>0 else np.full((准行-len(矩),参年+2),0) for 矩 in 月周股] 
for i in range(3): print((齐次周股[i])) 
def 热图(矩, idx): ax=sns.heatmap(矩[:,:6].astype(float),yticklabels=矩[:,-1],cmap='viridis',annot=True,fmt='.0f'); ax.set_title(f'{月份[idx]}',fontsize=10) 
for k in r(72): (k%6==0 and plt.figure(figsize=(14,12))); plt.subplot(2,3,k%6+1); 热图(齐次周股[k],k); (k%6==5 and plt.show()) # print(月内周股[i][:21])# 每期限幅20 
```
